```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np

r = requests.get("https://pubmed.ncbi.nlm.nih.gov/36473651/")
soup = bs(r.content)

all_titles = list()
all_hrefs = list()
all_dates = list()
all_authors = list()
all_abstracts = list()
    
for page_idx in range(0, 690):

    r = requests.get("https://pubmed.ncbi.nlm.nih.gov/?term=%22Cochrane%20Database%20syst%20rev%22%5BJournal%5D&filter=years.2014-2024&sort=date" 
                     + "&page=" + str(page_idx))
    soup = bs(r.content)
    
    divs = soup.find_all("div", {"class":"docsum-wrap"})
    for div in divs:
        title = div.find("div", {"class":"docsum-content"}).find("a", {"class":"docsum-title"}).text.strip()
        href = div.find("div", {"class":"docsum-content"}).find("a", {"class":"docsum-title"})["href"].strip()
        date = div.find("span", {"class":"docsum-journal-citation full-journal-citation"}).text.strip()
        full_authors = div.find("span", {"class":"docsum-authors full-authors"}).text.strip()
    
        new_link = "https://pubmed.ncbi.nlm.nih.gov/" + href
        r = requests.get(new_link)
        soup = bs(r.content)
    
        abstract_div = soup.find("div", {"class":"abstract"})
        try:
            abstract_strings = list(abstract_div.stripped_strings)
        except:
            abstract_strings = ['']
        
        all_titles.append(title)
        all_hrefs.append(href)
        all_dates.append(date)
        all_authors.append(full_authors)
        all_abstracts.append(abstract_strings)
    
    
    soup.find("title").text.strip()

```
